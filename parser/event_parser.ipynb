```python
from sqlalchemy import create_engine, text
import os
from dotenv import load_dotenv

engine = create_engine(f'postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@localhost:5432/f1pace')
```

```python
from fastf1 import get_event_schedule
import pandas as pd

events_schedule_df = pd.DataFrame()
for year in range(2018, 2027):

    response_frame = get_event_schedule(year)
    response_frame['year'] = year

    events_schedule_df = pd.concat([events_schedule_df, response_frame])
    
events_schedule_df
```

# Events


```python
events_df = (
    events_schedule_df[["year", "RoundNumber", "EventFormat", "Country", "Location"]]
    .reset_index(drop=True)
)
events_df["EventFormat"] = events_df["EventFormat"].apply(lambda x : False if x=="conventional" else True)
events_df
```

```python
circuits_df = pd.read_sql('SELECT id, country, location, circuit_name FROM tracks', engine)
circuits_df.head()
```

```python
circuits_df["circuit_info"] = (circuits_df["country"] + " " + circuits_df["location"] + " " + circuits_df["circuit_name"]).str.lower()
circuits_df[["id", "circuit_info"]].head()
```

```python
def to_circuit_id(row):

    location = row["Location"].lower()

    replace_dict = {
        "Montréal" : "Montreal", 
        "Yas Island" : "Yas", 
        "Miami Gardens" : "Miami",
    }
    for key in replace_dict:
        location = location.replace(key.lower(), replace_dict[key].lower())

    for i in range(len(circuits_df)):
        if location in circuits_df.loc[i, "circuit_info"]:
            return int(circuits_df.loc[i, "id"])
    return None

events_df["circuit_id"] = events_df.apply(to_circuit_id, axis=1)
events_df['circuit_id'] = events_df['circuit_id'].astype('Int32')
events_df
```

```python
events_df[pd.isna(events_df["circuit_id"])]
```

```python

events_df_clean = (
    events_df[events_df["RoundNumber"]>0][["circuit_id", "year", "RoundNumber", "EventFormat"]]
    .rename(columns={
        "RoundNumber" : "round",
        "EventFormat" : "is_sprint"
    })
)

events_df_clean
```

```python
events_df_clean.iloc[43]
```

```python
with engine.connect() as conn:
    conn.execute(text("DELETE FROM events"))
    conn.commit()
    
```

```python
events_df_clean.to_sql("events", engine, if_exists="append", index=True, index_label="id")
```

# Sessions

```python
session_names = ["Unknown",
                 "Practice 1", 
                 "Practice 2", 
                 "Practice 3", 
                 "Qualifying", 
                 "Race", 
                 "Sprint Qualifying", 
                 "Sprint"]

session_types_df = pd.DataFrame(session_names, columns=["name"])
session_types_df
```

```python
session_types_df.to_sql("session_types", engine, if_exists='append', index=True, index_label="id")
```

```python
session_data = []
for i in range(len(events_schedule_df)):
    event = events_schedule_df.iloc[i]

    if event["EventFormat"] == 'testing':
        continue

    event_id = events_df_clean[(events_df_clean["year"] == event["year"]) &
                               (events_df_clean["round"] == event["RoundNumber"])].index
     
    if len(event_id) != 1:
        print(event_id)
        raise ValueError
    event_id = event_id[0]

    for session_num in range(1, 6):
        s_type = event[f"Session{session_num}"]
        if s_type in session_names:
            s_type_ind = session_names.index(s_type)
        else:
            s_type_ind = session_names.index("Unknown")
        s_datetime = event[f"Session{session_num}DateUtc"]

        session_data.append((event_id, s_type_ind, s_datetime))
```

```python
sessions_df = pd.DataFrame(data=session_data, 
                        columns=["event_id", "session_type", "datetime_start_utc"])

sessions_df
```

```python
sessions_df.to_sql("sessions", engine, if_exists='append', index=True, index_label="id")
```
